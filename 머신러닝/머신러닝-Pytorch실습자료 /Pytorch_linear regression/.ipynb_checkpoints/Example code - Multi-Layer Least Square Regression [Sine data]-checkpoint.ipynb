# Least Square Regression - Multi-Layer Case
## (for sine wave data)

        layer 1                                   layer2
    x-> weight matrix1->non-linear activation->weight matrix2
         bias 1                                    bias2

# Library

```python
import numpy as np

import torch
import torch.nn as nn # for neural network models
import torch.optim as optim # for optimization 
import torch.nn.init as init # for initialization 

from matplotlib import pyplot as plt
```

# simple data

```python
# true data
N = 100
X = np.linspace(-np.pi, np.pi, N).reshape(N, 1)#N*1
real_y = np.sin(X)

# add noise
y = real_y + 0.1*np.random.randn(N, 1)

# shuffle
idx = np.random.permutation(N)
X_ = X[idx]
y_ = y[idx]
```

```python
# show
plt.figure()
plt.plot(X, real_y)
plt.scatter(X, y)
plt.show()
```

# Split train / test dataset

```python
n = N-20

X_train = X_[:n]
y_train = y_[:n]
X_test = X_[n:]
y_test = y_[n:]
```

```python
plt.figure()
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend()
plt.show()
```

# Multi-layer Linear Regression

hyperparameter: 사용자가 validation dataset을 통해 찾아야 하는 파라미터 값들

1.epoch
2.layer의 개수
3.각각의 layer의 weight의 크기

## Hyper parameter setting

```python
learning_rate = 1e-1
epochs = 1000 #epoch 전체 데이터를 한번 다 보는 것
#iteration: batch를 한번 보는 것
```

## Make model with pytorch (Gradient Descent)

```python
# prepare data
X_train = torch.tensor(X_train, dtype = torch.float) # if X_train is numpy.array
label = torch.tensor(y_train, dtype = torch.float)

# prepare model

#nn.Linear():neural network에서 layer를 나타냅니다
#(weight matrix, bias vector)
#nn.Linear(matrix의 row의 개수, column의 개수)

model = nn.Sequential(
        nn.Linear(1,1), #layer1
        nn.Sigmoid(),   #non-linear activation
        nn.Linear(1,1), #layer2
        )

# prepare optimization and loss function
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []
# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = torch.tensor(X_test, dtype = torch.float)
X_ = torch.tensor(X_, dtype = torch.float)

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```

# Change hidden layer's node number

```python
# prepare data
X_train = X_train.clone().detach() # if X_train is torch.tensor
label = label.clone().detach()

# prepare model
model = nn.Sequential(
        nn.Linear(1,10),
        nn.Sigmoid(),
        nn.Linear(10,1),
        )

# prepare optimziation and loss function 
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []

# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = X_test.clone().detach()
X_ = X_.clone().detach()

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```

# Change Activation function from sigmoid to tanh

### Activation func
1. sigmoid
2. tanh
3. ReLu

```python
# prepare data
X_train = X_train.clone().detach() # if X_train is torch.tensor
label = label.clone().detach()

# prepare model
model = nn.Sequential(
        nn.Linear(1,10),
        nn.Tanh(),
        nn.Linear(10,1),
        )

# prepare optimziation and loss function 
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []

# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = X_test.clone().detach()
X_ = X_.clone().detach()

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```

# Sum-up

 ## hyper parameter setting 
 learning_rate = 1e-4, 1e-3
 epochs 
 early stopping 모델이 학습할 때 valid의 grad가 더이상 줄어들지 않을 때 사용
 

nn.Linear() <- row 의 개수, col 개수
nn.Linear() : 하나의 layer를 의미

model(data)
:내부적으로 data*weight matrix+bias

model = nn.Linear(1,1)

output = model(data)

-> [1,1]


```python
#################### hyper parameter setting ####################
learning_rate=1e-1
epochs = 1000

h = 10 ### number of hidden layer nodes
act = nn.Tanh() ### activation function
#################################################################
# prepare data
X_train = torch.tensor(X_train, dtype = torch.float)
label = torch.tensor(y_train, dtype = torch.float)

# prepare model
model = nn.Sequential(
        nn.Linear(1,h),
        act,
        nn.Linear(h,1),
        )

# prepare optimziation and loss function 
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []

# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = torch.tensor(X_test, dtype = torch.float)
X_ = torch.tensor(X_, dtype = torch.float)

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```
