```python
dt1.공급일자 = pd.to_datetime(dt1['공급일자'])
dt2.공급일자 = pd.to_datetime(dt2['공급일자'])
dt3.공급일자 = pd.to_datetime(dt3['공급일자'])
dt4.공급일자 = pd.to_datetime(dt4['공급일자'])
```

```python
import datetime
t = ['월','화','수','목','금','토','일']
```

```python
dt1['월'] = dt1.공급일자.dt.month
dt2['월'] = dt2.공급일자.dt.month
dt3['월'] = dt3.공급일자.dt.month
dt4['월'] = dt4.공급일자.dt.month
```

```python
dt1['요일'] = dt1.공급일자.dt.weekday
dt2['요일'] = dt2.공급일자.dt.weekday
dt3['요일'] = dt3.공급일자.dt.weekday
dt4['요일'] = dt4.공급일자.dt.weekday
```

```python
dt1.요일 = dt1.요일.apply(lambda x : t[x])
dt2.요일 = dt2.요일.apply(lambda x : t[x])
dt3.요일 = dt3.요일.apply(lambda x : t[x])
dt4.요일 = dt4.요일.apply(lambda x : t[x])
```

```python
#수량=0=금액 제거 코드
dt1_1 = dt1.copy()
a=dt1_1[(dt1_1["구매금액"]==0)&(dt1_1["구매수량"]==0)].index
dt1_1.drop(a,axis=0,inplace=True)
#구매수량 -인데 구매이력이 없고 구매금액이 0인것 제거
dt1_1.drop(79299,axis=0,inplace=True)
#통삼겹이랑 삼겹살 제거
dt1_1.drop(17988,axis=0,inplace=True)
dt1_1.drop(109050,axis=0,inplace=True)
dt1_1.reset_index(drop=True,inplace=True) 

dt2_1 = dt2.copy()
a=dt2_1[(dt2_1["구매금액"]==0)&(dt2_1["구매수량"]==0)].index
dt2_1.drop(a,axis=0,inplace=True)
dt2_1.reset_index(drop=True,inplace=True) 

dt3_1 = dt3.copy()
a=dt3_1[(dt3_1["구매금액"]==0)&(dt3_1["구매수량"]==0)].index
dt3_1.drop(a,axis=0,inplace=True)
dt3_1.reset_index(drop=True,inplace=True)

dt4_1 = dt4.copy()
a=dt4_1[(dt4_1["구매금액"]==0)&(dt4_1["구매수량"]==0)].index
dt4_1.drop(a,axis=0,inplace=True)
dt4_1.drop([25602, 51584], axis = 0, inplace = True)  # 구매 없이 환불된 기록 누락
dt4_1.loc[51513,'반품_원거래일자'] = 20181025 # 반품 날짜 오류 수정
dt4_1.reset_index(drop=True,inplace=True) 
```
