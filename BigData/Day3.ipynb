# 의사결정나무

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('체질검사.csv', engine = 'python')
```

```python
df_y = df['FAT']
df_x = df.drop('FAT', axis = 1, inplace = False)

df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_x, df_y, test_size = 0.3, random_state = 1234)
```

```python
print("train data X size : {}".format(df_train_x.shape))
print("train data Y size : {}".format(df_train_y.shape))
print("test data X size : {}".format(df_test_x.shape))
print("test data Y size : {}".format(df_test_y.shape))
```

```python
tree_uncustomized = DecisionTreeRegressor(random_state = 1234)
tree_uncustomized.fit(df_train_x, df_train_y)

print("Score on training set : {:.3f}".format(tree_uncustomized.score(df_train_x, df_train_y)))
print("Score on test set : {:.3f}".format(tree_uncustomized.score(df_test_x, df_test_y)))
```

## min_samples_leaf Parameter 조정

```python
train_score = [];test_score= []
para_leaf = [n_leaf * 1 for n_leaf in range(1,20)]

for v_min_samples_leaf in para_leaf:
    tree = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = v_min_samples_leaf)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x,df_train_y))
    test_score.append(tree.score(df_test_x,df_test_y))
    
    
df_score_leaf = pd.DataFrame()
df_score_leaf['MinSamplesLeaf'] = para_leaf
df_score_leaf['TrainScore'] = train_score
df_score_leaf['TestScore'] = test_score
    
```

```python
df_score_leaf.round(3)
```

```python
plt.plot(para_leaf, train_score, linestyle = '-', label = "Train Score")
plt.plot(para_leaf, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

```python
v_feature_name = df_train_x.columns
tree_graph = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = 8)
tree_graph.fit(df_train_x, df_train_y)
export_graphviz(tree_graph, out_file = 'tree_leaf.dot',
                feature_names = v_feature_name, impurity = True, filled = True)
with open("tree_leaf.dot") as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

## min_samples_split Parameter 조정

```python
train_score = [];test_score= []
para_split = [n_split * 2 for n_split in range(2,20)]

for v_min_samples_split in para_split:
    tree = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = 8, 
                                 min_samples_split = v_min_samples_split)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x,df_train_y))
    test_score.append(tree.score(df_test_x,df_test_y))
    
    
df_score_split = pd.DataFrame()
df_score_split['MinSamplesSplit'] = para_split
df_score_split['TrainScore'] = train_score
df_score_split['TestScore'] = test_score
df_score_split.round(3)
```

```python
plt.plot(para_split, train_score, linestyle = '-', label = "Train Score")
plt.plot(para_split, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

#### 분리 노드의 최소 자료 수 변경 결과 비교(20vs26)

```python
tree_graph = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = 8, min_samples_split = 20)
tree_graph.fit(df_train_x, df_train_y)
export_graphviz(tree_graph, out_file = "tree_split.dot",
               feature_names = v_feature_name, impurity=True, filled = True)
with open("tree_split.dot") as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

## max_depth Parameter 조정

```python
train_score = [];test_score= []
para_depth = [depth for depth in range(1,11)]

for v_max_depth in para_depth:
    tree = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = 8, 
                                 min_samples_split = 20, max_depth = v_max_depth)
    tree.fit(df_train_x, df_train_y)
    train_score.append(tree.score(df_train_x,df_train_y))
    test_score.append(tree.score(df_test_x,df_test_y))
    
    
df_score_depth = pd.DataFrame()
df_score_depth['MinSamplesSplit'] = para_depth
df_score_depth['TrainScore'] = train_score
df_score_depth['TestScore'] = test_score
df_score_depth.round(3)
```

```python
plt.plot(para_depth, train_score, linestyle = '-', label = "Train Score")
plt.plot(para_depth, test_score, linestyle = '--', label = "Test Score")
plt.legend()
```

```python
tree_graph = DecisionTreeRegressor(random_state = 1234, min_samples_leaf = 8, min_samples_split = 20, max_depth = 4)
tree_graph.fit(df_train_x, df_train_y)
export_graphviz(tree_graph, out_file = "tree_split.dot",
               feature_names = v_feature_name, impurity=True, filled = True)
with open("tree_split.dot") as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

## 최종 모델 선정

```python
tree_final = DecisionTreeRegressor(min_samples_leaf = 8, min_samples_split = 20, max_depth = 4, random_state= 1234)
tree_final.fit(df_train_x, df_train_y)
export_graphviz(tree_final, out_file = "tree_split.dot",
               feature_names = v_feature_name, impurity=True, filled = True)
with open("tree_split.dot") as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

## 설명변수의 중요도

```python
df_importance = pd.DataFrame()
df_importance['Feature']=v_feature_name
df_importance['Importance'] = tree_final.feature_importances_

df_importance.sort_values("Importance",ascending = False, inplace = True)
df_importance.round(3)
```

```python
df_importance.sort_values("Importance", ascending = True, inplace =True)
coordinates = range(len(df_importance))
plt.barh(y = coordinates, width = df_importance["Importance"])
plt.yticks(coordinates, df_importance["Feature"])
plt.xlabel("변수 중요도")
plt.ylabel("변수")
```

## GridSearchCV

```python
estimator = DecisionTreeRegressor()

param_grid = {"criterion":["mse","friedman_mse","mae"], "max_features":["auto", "sqrt","log2"],
             "max_depth":para_depth, "min_samples_split":para_split, "min_samples_leaf":para_leaf}

grid_dt = GridSearchCV(estimator, param_grid, scoring = "r2", n_jobs = -1)
grid_dt.fit(df_train_x, df_train_y)

print("best estimator model:\n{}".format(grid_dt.best_estimator_))
print("\nbest parameter:\n{}".format(grid_dt.best_params_))
print("\nbest score:\n{}".format(grid_dt.best_score_.round(3)))
```

# 앙상블

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('체질검사.csv', engine = 'python')
df_y = df['FAT']
df_x = df.drop('FAT', axis = 1, inplace = False)

df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_x, df_y, test_size = 0.3, random_state = 1234)
```

```python
rf_uncustomized = RandomForestRegressor(random_state = 1234)
rf_uncustomized.fit(df_train_x, df_train_y)

print("Score on training set : {:.3f}".format(rf_uncustomized.score(df_train_x, df_train_y)))
print("Score on test set : {:.3f}".format(rf_uncustomized.score(df_test_x, df_test_y)))
```

## n_estimators Parameter 조정

```python
train_score = [];test_score= []
para_n_tree = [n_tree * 10 for n_tree in range(1,15)]

for v_n_estimators in para_n_tree:
    rf = RandomForestRegressor(random_state = 1234, n_estimators = v_n_estimators)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x,df_train_y))
    test_score.append(rf.score(df_test_x,df_test_y))
    
    
df_score_n = pd.DataFrame()
df_score_n['n_estimators'] = para_n_tree
df_score_n['TrainScore'] = train_score
df_score_n['TestScore'] = test_score
df_score_n.round(3)
```

```python
plt.plot(para_n_tree, train_score, linestyle='-', label = "Train Score")
plt.plot(para_n_tree, test_score, linestyle='--', label = "Test Score")
plt.legend()
```

## min_samples_leaf Parameter 조정

```python
train_score = [];test_score= []
para_leaf = [n_leaf * 1 for n_leaf in range(1,21)]

for v_min_samples_leaf in para_leaf:
    rf = RandomForestRegressor(random_state = 1234, n_estimators = 130,
                                 min_samples_leaf = v_min_samples_leaf)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x,df_train_y))
    test_score.append(rf.score(df_test_x,df_test_y))
    
    
df_score_leaf = pd.DataFrame()
df_score_leaf['MinSamplesLeaf'] = para_leaf
df_score_leaf['TrainScore'] = train_score
df_score_leaf['TestScore'] = test_score
df_score_leaf
```

```python
plt.plot(para_leaf, train_score, linestyle='-', label = "Train Score")
plt.plot(para_leaf, test_score, linestyle='--', label = "Test Score")
plt.legend()
```

## min_samples_split Parameter 조정

```python
train_score = [];test_score= []
para_split = [n_split * 2 for n_split in range(2,21)]

for v_min_samples_split in para_split:
    rf = RandomForestRegressor(random_state = 1234, n_estimators = 130,min_samples_leaf = 6, 
                                 min_samples_split = v_min_samples_split)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x,df_train_y))
    test_score.append(rf.score(df_test_x,df_test_y))
    
    
df_score_split = pd.DataFrame()
df_score_split['MinSamplesSplit'] = para_split
df_score_split['TrainScore'] = train_score
df_score_split['TestScore'] = test_score
df_score_split.round(3)
```

```python
plt.plot(para_split, train_score, linestyle='-', label = "Train Score")
plt.plot(para_split, test_score, linestyle='--', label = "Test Score")
plt.legend()
```

## depth Parameter 조정

```python
train_score = [];test_score= []
para_depth = [depth for depth in range(1,11)]

for v_max_depth in para_depth:
    rf = RandomForestRegressor(random_state = 1234, n_estimators = 130, min_samples_leaf = 6, 
                                 min_samples_split = 14, max_depth = v_max_depth)
    rf.fit(df_train_x, df_train_y)
    train_score.append(rf.score(df_train_x,df_train_y))
    test_score.append(rf.score(df_test_x,df_test_y))
    
    
df_score_depth = pd.DataFrame()
df_score_depth['Depth'] = para_depth
df_score_depth['TrainScore'] = train_score
df_score_depth['TestScore'] = test_score
df_score_depth.round(3)
```

```python
plt.plot(para_depth, train_score, linestyle='-', label = "Train Score")
plt.plot(para_depth, test_score, linestyle='--', label = "Test Score")
plt.legend()
```

## 최종 모델 생성

```python
v_feature_name = df_train_x.columns
rf_final = RandomForestRegressor(n_estimators = 130, min_samples_leaf = 6, min_samples_split = 14, max_depth = 5, random_state= 1234)
rf_final.fit(df_train_x, df_train_y)

print("Score on training set: {:.3f}".format(rf_final.score(df_train_x, df_train_y)))
print("Score on test set: {:.3f}".format(rf_final.score(df_test_x, df_test_y)))
```
