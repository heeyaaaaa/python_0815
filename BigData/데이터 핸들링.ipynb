# 데이터 선택 filter / mask

```python
df = pd.DataFrame({'ABC':['A','B','C','A'],'abc':['a','b','c','c'], 
                   'group1':['G1','G2','G3','G4'],'value1':[11,13,12,14],'val1_mod':[0.231,0.197,0.213,0.199]})
df
```

```python
df.filter(items=['group1','value1'])
```

```python
df.filter(regex='1',axis= 1)
```

```python
df.filter(regex='1$',axis = 1)
```

```python
df.filter(like='a', axis = 1)
```

```python
df.filter(regex='[Aa]', axis = 1)
```

```python
df.mask(df.value1>12)
```

```python
df.mask(df.abc == 'a')
```

```python
df.mask(cond = df.value1>12, other = 999)
```

# 데이터 결합

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']}, index = [0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],'D':['D4','D5','D6','D7']}, index = [4,5,6,7])
df2
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],'D':['D8','D9','D10','D11']}, index = [8,9,10,11])
df3
```

```python
result = pd.concat([df1, df2, df3])
result
```

```python
result = pd.concat([df1,df2,df3], axis = 1, ignore_index=True)
result
```

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                    'key2':['K0','K1','K0','K1'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                    'key2':['K0','K0','K0','K0'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
result = pd.merge(left,right, on = ['key1','key2'])
result
```

```python
result = pd.merge(left,right, on = ['key1','key2'], how = 'left')
result
```

```python
result = pd.merge(left,right, on = ['key1','key2'], how = 'right')
result
```

```python
result = pd.merge(left,right, on = ['key1','key2'], how = 'outer')
result
```

```python
result = pd.merge(left,right, on = ['key1','key2'], how = 'outer', indicator=True)
result
```

```python
left = pd.DataFrame({'key1':[1,2,3,4],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'C':['C1','C2','C5'],
                    'D':['D1','D2','D5']}, index = [1,2,5])
```

```python
result = left.join(right)
result
```

```python
result = left.join(right, how = 'right')
result
```

```python
result = left.join(right, how = 'inner')
result
```

```python
result = left.join(right, how = 'outer')
result
```

```python
result = left.join(right, on = 'key1')
result
```

```python
result2 = pd.merge(left, right, left_on = 'key1', right_index = True, how = 'left')
result2
```

```python
df = pd.DataFrame({'factor1':['A','A','A','A','A','B','B','B','B'],
                  'factor2':['a','b','c','a','b','a','b','c','c'],
                   'value':[11,13,12,12,14,12,13,11,12]})
```

```python
df
```

```python
pd.crosstab(df.factor1, df.factor2)
```

```python
pd.crosstab(df.factor1, df.factor2, rownames = ['Fact1'], colnames = ['Fact2'],margins = True)
```

```python
pd.crosstab(df.factor1, [df.factor2, df.value])
```

```python
pd.crosstab(df.factor1, df.factor2, margins = True, normalize = True)
```
